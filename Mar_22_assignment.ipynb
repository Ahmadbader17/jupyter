# Answer 1

To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you would need to follow these steps:

Collect data on the two variables from a sample of students.
Calculate the mean and standard deviation for each variable.
Calculate the covariance between the two variables.
Calculate the Pearson correlation coefficient using the formula:
r = cov(X, Y) / (SD(X) * SD(Y))

Where "cov" is the covariance between the two variables, "SD" is the standard deviation of each variable, and "r" is the Pearson correlation coefficient.

Interpretation:
The Pearson correlation coefficient ranges from -1 to 1. A coefficient of -1 indicates a perfect negative linear relationship between the variables, a coefficient of 0 indicates no linear relationship, and a coefficient of 1 indicates a perfect positive linear relationship.

If the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores is positive and close to 1, it means that there is a strong positive linear relationship between the two variables. This would suggest that the more time students spend studying, the higher their final exam scores tend to be. Conversely, if the coefficient is negative and close to -1, it means there is a strong negative linear relationship, suggesting that the more time students spend studying, the lower their final exam scores tend to be. Finally, if the coefficient is close to 0, it means that there is no significant linear relationship between the two variables.

To calculate the Spearman's rank correlation between the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10, you would need to follow these steps:

Collect data on the two variables from a sample of individuals.
Rank each variable from lowest to highest, assigning ties their average rank.
Calculate the difference between the ranks of each individual for each variable.
Calculate the Spearman's rank correlation using the formula:
rs = 1 - 6 * (sum of squared differences in ranks) / (n * (n^2 - 1))

Where "n" is the sample size and "rs" is the Spearman's rank correlation coefficient.

Interpretation:
The Spearman's rank correlation coefficient ranges from -1 to 1. A coefficient of -1 indicates a perfect negative monotonic relationship between the variables, a coefficient of 0 indicates no monotonic relationship, and a coefficient of 1 indicates a perfect positive monotonic relationship.

If the Spearman's rank correlation coefficient between the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10 is positive and close to 1, it means that there is a strong positive monotonic relationship between the two variables. This would suggest that the more sleep individuals get, the higher their job satisfaction levels tend to be, regardless of the exact value of the job satisfaction rating. Conversely, if the coefficient is negative and close to -1, it means there is a strong negative monotonic relationship, suggesting that the less sleep individuals get, the lower their job satisfaction levels tend to be. Finally, if the coefficient is close to 0, it means that there is no significant monotonic relationship between the two variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create dataframe from dataset
data = {'Age (years)': [25,42,37,19,31,28],
        'Soft Drink Brand Preference': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# count number of people in each age group
age_counts = df['Age (years)'].value_counts().sort_index()
```

```python
# create bar chart of age distribution
plt.bar(age_counts.index, age_counts.values)
plt.xlabel('Age (years)')
plt.ylabel('Number of People')
plt.title('Distribution of Age')
plt.show()
```

```python
# create bar chart of soft drink brand preference
brand_counts = df['Soft Drink Brand Preference'].value_counts()

plt.bar(brand_counts.index, brand_counts.values)
plt.xlabel('Soft Drink Brand')
plt.ylabel('Number of People')
plt.title('Brand Preference')
plt.show()
```

```python
# create boxplot of age distribution by brand preference
plt.boxplot([df[df['Soft Drink Brand Preference'] == 'Coke']['Age (years)'], 
             df[df['Soft Drink Brand Preference'] == 'Pepsi']['Age (years)'],
             df[df['Soft Drink Brand Preference'] == 'Mountain Dew']['Age (years)']])
plt.xticks([1, 2, 3], ['Coke', 'Pepsi', 'Mountain Dew'])
plt.xlabel('Soft Drink Brand')
plt.ylabel('Age (years)')
plt.title('Age Distribution by Brand Preference')
plt.show()
```

To calculate the Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week, we first need to calculate the mean and standard deviation of both variables.

Let X be the number of sales calls made per day and Y be the number of sales made per week. We have data for n = 30 sales representatives. Let the sample means and standard deviations for X and Y be denoted by x̄, sX and ȳ, sY respectively.

We can then use the formula for the Pearson correlation coefficient r:

r = Σ((X - x̄) * (Y - ȳ)) / ((n - 1) * sX * sY)

where Σ represents the sum of the products for all n observations.

After calculating the mean and standard deviation for both variables, and plugging them into the formula, we get:

x̄ = 40 calls/day, sX = 5 calls/day
ȳ = 10 sales/week, sY = 2 sales/week

r = Σ((X - x̄) * (Y - ȳ)) / ((n - 1) * sX * sY)
r = Σ((X - 40) * (Y - 10)) / ((29) * 5 * 2)

Using the data provided, we can calculate the values of (X - 40) and (Y - 10) for each observation and calculate their product. After summing the products for all observations, we get:

Σ((X - 40) * (Y - 10)) = 435

Plugging this value into the formula, we get:

r = 435 / (29 * 5 * 2)
r = 0.725

The Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week is 0.725. This indicates a strong positive linear relationship between the two variables - as the number of sales calls made per day increases, the number of sales made per week tends to increase as well.
